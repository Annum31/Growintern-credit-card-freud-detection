import libraries and dependencies

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

loading a dataset to a pandas dataframe

```python
credit_card_data= pd.read_csv('creditcard.csv')
```

```python
#first 5 rows of te dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#check the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fraudulant transaction
credit_card_data['Class'].value_counts()
```

this dataset is highly unbalanced

0.... normal transaction

1.... fraudulent transaction

```python
#separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for bothh transactions
credit_card_data.groupby('Class').mean()
```

under sampling
build a sample dataset containing similar distributuion of normal transaction and fraudulent transaction

Number of fraudulent transaction = 492

```python
legit_sample = legit.sample(n=492)
```

Concatinatingg two Dataframes

```python
new_dataset= pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and targets

```python
X= new_dataset.drop(columns='Class', axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training Data and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

logistic Regression

```python
model = LogisticRegression()
```

#Training thhe logistic regression model wit Training Data

```python
model.fit(X_train, Y_train)
```

Model Evaluation

Acuracy score

```python
#accuracy on training data
X_train_prediction= model.predict(X_train)
trainin_data_accuracy= accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data :' , trainin_data_accuracy)
```

```python
#accuracy on test Data
X_test_prediction= model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data:', test_data_accuracy)
```
